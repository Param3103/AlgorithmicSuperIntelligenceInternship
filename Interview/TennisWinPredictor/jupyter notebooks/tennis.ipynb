```python
# generate winner for each grand slam from 2000 onwards
import pandas as pd

df = pd.read_csv("../tennis_atp-master/atp_matches_2000.csv")

# Extract year
df["year"] = df["tourney_date"].astype(str).str[:4].astype(int)


grand_slams = [
    "Australian Open",
    "Roland Garros",
    "Wimbledon",
    "US Open"
]

# Keep only Grand Slam finals
gs_finals = df[(df["tourney_name"].isin(grand_slams)) & (df["round"] == "F")]

# Get winner of each grand slam each year
winners1 = gs_finals[["year", "tourney_name", "winner_name"]].sort_values(["year", "tourney_name"])

print(winners1)
```

```python
df = pd.read_csv("../tennis_atp-master/atp_matches_2001.csv")

# Extract year
df["year"] = df["tourney_date"].astype(str).str[:4].astype(int)


grand_slams = [
    "Australian Open",
    "Roland Garros",
    "Wimbledon",
    "US Open"
]

# Keep only Grand Slam finals
gs_finals = df[(df["tourney_name"].isin(grand_slams)) & (df["round"] == "F")]

# Get winner of each grand slam each year
winners2 = gs_finals[["year", "tourney_name", "winner_name"]].sort_values(["year", "tourney_name"])

print(winners2)
ans = pd.concat([winners1, winners2])
ans
```

```python

ans = []
for i in range(2000, 2025):
    df = pd.read_csv(f"../tennis_atp-master/atp_matches_{i}.csv")
    
    # Extract year
    df["year"] = df["tourney_date"].astype(str).str[:4].astype(int)


    grand_slams = [
        "Australian Open",
        "Roland Garros",
        "Wimbledon",
        "US Open"
    ]

    # Keep only Grand Slam finals
    gs_finals = df[(df["tourney_name"].isin(grand_slams)) & (df["round"] == "F")]

    # Get winner of each grand slam each year
    winners = gs_finals[["year", "tourney_name", "winner_name"]].sort_values(["year", "tourney_name"])
    ans.append(winners)
ans = pd.concat(ans)
ans
```

```python
# rows will be year, column will be tourney name, value will be winner
table = ans.pivot(index="year", columns="tourney_name", values="winner_name")

print(table)
```
